```
import numpy as np
import matplotlib.pyplot as plt
from ipywidgets import interact

# Define a function to plot the roots for specific values of u and v
def plot_roots(u, v):
    plt.figure(figsize=(6, 6))

    # Define coefficients of the equation x^4 - (2u^2 + v^2)x^2 + u^4 = 0
    coeff = [1, 0, -(2*u**2 + 2*v**2), 0, (u**4 - 2*u**2*v**2 + v**4)]


    # Solve the quartic equation
    roots = np.roots(coeff)

    # Extract real and imaginary parts of the roots
    real_parts = np.real(roots)
    imaginary_parts = np.imag(roots)

    # Plot the roots on the complex plane
    plt.scatter(real_parts, imaginary_parts, color='blue', s=50)

    # Set plot limits and labels
    plt.xlim(-2, 2)
    plt.ylim(-2, 2)
    plt.axhline(0, color='black', linewidth=0.5)
    plt.axvline(0, color='black', linewidth=0.5)

    # Add title and labels
    plt.title(rf"Eigenvalues of N=2 Closed Chain SSH for u={u:.2f}, v={v:.2f}")
    plt.xlabel('Real')
    plt.ylabel('Imaginary')
    plt.grid(True)

    # Show the plot
    plt.show()

# Use interact to create sliders for u and v
interact(plot_roots, u=(0, 1.0, 0.01), v=(0, 1.0, 0.01));
```
